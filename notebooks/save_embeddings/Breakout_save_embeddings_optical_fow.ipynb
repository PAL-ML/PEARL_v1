# Convert to embeddings

## Function definitions

```python
def get_embeddings(encoder, episodes, save_per_episode=None, data_dir=None, embeddings_filename=None):
    embeddings_list = []
    for ep in tqdm(range(len(episodes))):
        embeddings = []
        for img in episodes[ep]:
            inp = img.float().unsqueeze(0).cuda()/255.
            with torch.no_grad():
                feature = encoder(inp)
            embeddings.append(feature.squeeze(0).detach()*255)
        embeddings_list.append(embeddings)

        if embeddings_filename:
            if (ep+1)%save_per_episode == 0 or ep == len(episodes)-1:
                torch.save(embeddings_list, os.path.join(data_dir, embeddings_filename))

    return embeddings_list
```
