# Helper functions

```python
class SimclrMLPEncoder(nn.Module):
  def __init__(self, input_size, output_size, hidden_size=2048):
    super().__init__()
    self.device = "cuda" if torch.cuda.is_available() else "cpu"
    self.linear1 = nn.Linear(input_size, hidden_size)
    self.linear2 = nn.Linear(hidden_size, output_size)
    self.bn = nn.BatchNorm1d(hidden_size)
    self.feature_size =  output_size
  
  def forward(self, inputs):
    x = self.bn(F.relu(self.linear1(inputs)))
    out = self.linear2(x)
    return out
```

```python
def concat_embeddings_with_full_img(aug_eps,  full_eps):
  processed_eps = []
  for i, ep in enumerate(aug_eps):
    processed_ep = []
    for j, s in enumerate(ep):
        index_full =  j 
        processed_emb = torch.cat([aug_eps[i][index_full], full_eps[i][index_full]], dim=0)
        processed_ep.append(processed_emb)
    processed_eps.append(processed_ep)
  return processed_eps
```

```python
class CLIPSimCLRTrainer(Trainer):
    def __init__(self, encoder, config, device=torch.device('cpu'), wandb=None):
        super().__init__(encoder, wandb, device)
        self.config = config
        self.patience = self.config["patience"]
        self.epochs = config['epochs']
        self.batch_size = config['batch_size']
        self.save_dir = config['save_dir']
        self.lr = config['lr']
        self.weight_decay = config['weight_decay']
        self.early_stopper = EarlyStopping(patience=self.patience, verbose=False, wandb=self.wandb, name="encoder")
        self.name = config['model_name']
        self.temperature = config['temperature']
        self.n_views = 2 

        self.criterion = torch.nn.CrossEntropyLoss().to(device)
        self.optimizer = torch.optim.Adam(self.encoder.parameters(), self.lr, weight_decay=self.weight_decay)
    
    def reset_scheduler(self, steps):
        self.scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(self.optimizer, steps)

    def generate_batch(self, episodes_t1, episodes_t2):
      total_steps = sum([len(e) for e in episodes_t1])
      print('Total Steps: {}'.format(total_steps))
      # Episode sampler
      # Sample `num_samples` episodes then batchify them with `self.batch_size` episodes per batch
      sampler = BatchSampler(RandomSampler(range(len(episodes_t1)),
                              replacement=True, num_samples=total_steps),
                              self.batch_size, drop_last=True)

      for indices in sampler:
          episodes_batch_t1 = [episodes_t1[x] for x in indices]
          episodes_batch_t2 = [episodes_t2[x] for x in indices]

          x_t1, x_t2 = [], []
          for episode in episodes_batch_t1:
              i_t1 = np.random.randint(0, len(episode))
              x_t1.append(episode[i_t1])

          for episode in episodes_batch_t2:
              i_t2 = np.random.randint(0, len(episode))
              x_t2.append(episode[i_t2])

          yield torch.stack(x_t1).float().to(self.device) / 255., torch.stack(x_t2).float().to(self.device) / 255.
    
    def do_one_epoch(self, epoch, episodes_t1, episodes_t2):
        mode = "train" if self.encoder.training else "val"
        epoch_loss, accuracy, steps = 0., 0., 0
        epoch_loss = 0.
        data_generator = self.generate_batch(episodes_t1, episodes_t2)
        
        #total_steps = sum([len(e) for e in episodes_t1])
        #total_steps = total_steps
        #self.reset_scheduler(total_steps)

        for x_t1, x_t2 in data_generator:
            
            input_emb = torch.cat([x_t1, x_t2], dim=0)
            feats = self.encoder(input_emb)

            logits, labels = self.info_nce_loss(feats)

            loss = self.criterion(logits, labels)

            self.optimizer.zero_grad()
            if mode == "train":
                loss.backward()
                self.optimizer.step()
            
            epoch_loss += loss.detach().item()
            steps += 1
            #self.scheduler.step()
            
        self.log_results(epoch, epoch_loss / steps, prefix=mode)
        if mode == "val":
            self.early_stopper(-epoch_loss / steps, self.encoder)
    
    def train(self, tr_eps_t1, tr_eps_t2, val_eps_t1, val_eps_t2, save_interval=5):

        for e in range(self.epochs):
            self.encoder.train()
            self.do_one_epoch(e, tr_eps_t1, tr_eps_t2)

            self.encoder.eval()
            self.do_one_epoch(e, val_eps_t1, val_eps_t2)
            
            if e % save_interval == 0 and e != 0:
                self.save_checkpoint(self.name, self.encoder, num_epochs=e)

            if self.early_stopper.early_stop:
                break 

        self.save_checkpoint(self.name, self.encoder)

    def info_nce_loss(self, features):

        labels = torch.cat([torch.arange(self.batch_size) for i in range(self.n_views)], dim=0)
        labels = (labels.unsqueeze(0) == labels.unsqueeze(1)).float()
        labels = labels.to(device)

        features = F.normalize(features, dim=1)

        similarity_matrix = torch.matmul(features, features.T)

        # discard the main diagonal from both: labels and similarities matrix
        mask = torch.eye(labels.shape[0], dtype=torch.bool).to(self.device)
        labels = labels[~mask].view(labels.shape[0], -1)
        similarity_matrix = similarity_matrix[~mask].view(similarity_matrix.shape[0], -1)

        # select and combine multiple positives
        positives = similarity_matrix[labels.bool()].view(labels.shape[0], -1)

        # select only the negatives the negatives
        negatives = similarity_matrix[~labels.bool()].view(similarity_matrix.shape[0], -1)

        logits = torch.cat([positives, negatives], dim=1)
        labels = torch.zeros(logits.shape[0], dtype=torch.long).to(self.device)

        logits = logits / self.temperature

        return logits, labels

    def save_checkpoint(self, name, model, num_epochs=None):
        '''Saves model'''
        
        if num_epochs:
            filepath = str(self.save_dir) + "/" + str(name) + '_' + str(num_epochs) + ".pt"
        else:
            filepath = str(self.save_dir) + "/" + str(name) + '_' + "final" + ".pt"

        torch.save(model.state_dict(), filepath)
    
    def log_results(self, epoch_idx, epoch_loss, prefix=""):
        print("{} Epoch: {}, Epoch Loss: {}, {}".format(prefix.capitalize(), epoch_idx, epoch_loss,
                                                                     prefix.capitalize()))
        self.wandb.log({prefix + '_loss': epoch_loss}, step=epoch_idx)
```
